```python
#Filling all the null values with NaN
df.fillna("NaN", inplace=True)
```

```python
#Checking if filling null values with NaN worked
df.isnull().sum()
```

```python
#Changing data type of ORDER_CREATION_DATE from integer to date time
df['ORDER_CREATION_DATE'] = pd.to_datetime(df['ORDER_CREATION_DATE'], format='%Y%m%d')
```

```python
#Changing data type of REQUESTED_DELIVERY_DATE from integer to date time
df['REQUESTED_DELIVERY_DATE'] = pd.to_datetime(df['REQUESTED_DELIVERY_DATE'], format='%Y%m%d')
```

```python
#Checking if it is displaying as date time
df
```

```python
#Checking if datatype changed
df.info()
```

```python
#Sanity check: checking if the order is valid based on creation date and delivery date
sanity_check = df[df['ORDER_CREATION_DATE'] > df['REQUESTED_DELIVERY_DATE']]
len(sanity_check)
```

```python
#Removing recored where order creation date is greater than delivery date
df = df[df['ORDER_CREATION_DATE'] <= df['REQUESTED_DELIVERY_DATE']]

#Checking whether the records are removed
sanity_check = df[df['ORDER_CREATION_DATE'] > df['REQUESTED_DELIVERY_DATE']]
len(sanity_check)
```

```python
#Checking the number of records with '-' in the feature ORDER_AMOUNT
count = df['ORDER_AMOUNT'].str.contains('-').sum()
print("Number of records with '-' in ORDER_AMOUNT:", count)
```

```python
#Replacing '-' with '' in the feature ORDER_AMOUNT
df['ORDER_AMOUNT'] = df.ORDER_AMOUNT.apply(lambda x: x.replace('-',''))
```

```python
#Checking whether '-' is replaces with '' in the feature ORDER_AMOUNT
count = df['ORDER_AMOUNT'].str.contains('-').sum()
print("Number of records with '-' in ORDER_AMOUNT:", count)
```

```python
#Replacing ',' with '.' in the feature ORDER_AMOUNT
df['ORDER_AMOUNT'] = df.ORDER_AMOUNT.apply(lambda x: x.replace(',','.'))
```

```python
#Checking whether ',' is replaces with '.' in the feature ORDER_AMOUNT
count = df['ORDER_AMOUNT'].str.contains(',').sum()
print("Number of records with ',' in ORDER_AMOUNT:", count)
```

```python
#Number of records where order date and delivery date is same
same_date = df[df['ORDER_CREATION_DATE'] == df['REQUESTED_DELIVERY_DATE']]
len(same_date)
```

```python
#Number of records for each type of currency
currency_counts = df['ORDER_CURRENCY'].value_counts()
print(currency_counts)
```

```python
#New column in the existing dataframe created as “'amount_in_usd'” and all 
#the non-USD currencies converted in USD and stored them in the same column.
exchange_rates = {
    'USD':1,                                                     
    'EUR': 1.08,   
    'AUD': 0.65,    
    'CAD': 0.74,    
    'GBP': 1.24,    
    'MYR': 0.22,    
    'PLN': 0.24,    
    'AED': 0.27,   
    'HKD': 0.13,     
    'CHF': 1.11,     
    'RON': 0.22,     
    'SGD': 0.74,     
    'CZK': 0.045,     
    'HU1': 0.0029,     
    'NZD': 0.61,       
    'BHD': 2.65,      
    'SAR': 0.27,       
    'QAR': 0.27,       
    'KWD': 3.25,       
    'SEK': 0.094
}
df['amount_in_usd'] = df['ORDER_AMOUNT'].astype(float) * df['ORDER_CURRENCY'].map(exchange_rates)
```

```python
df
```

```python
#Checking for zero values in 'amount_in_usd' column
zero_values = df[df['amount_in_usd'] == 0]
print(zero_values)
print("Hence",len(zero_values),"number of zeroes are there")
```

```python
df['Unique_Cust_ID'] = df['CUSTOMER_NUMBER'] + df['COMPANY_CODE'] 
```
